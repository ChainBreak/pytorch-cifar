```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as fn
import matplotlib.pyplot as plt
from pathlib import Path

%matplotlib notebook
```

```python
p = {}
p["results"] = {}
```

```python
p["checkpoint_path"] = Path("../checkpoint")
p["model_name"] = "ResNet"
p["end_epoch"] = 100
```

```python
def load_epoch(epoch,p): 
    path = p["checkpoint_path"] / p["model_name"] / f"{epoch:05}.pkl"
    return torch.load(path) 
```

```python

end_checkpoint = load_epoch(p["end_epoch"],p)
results = {}
for epoch in range(p["end_epoch"]+1):
    print(epoch)
    epoch_checkpoint = load_epoch(epoch,p)
    
    for param_name in end_checkpoint.keys():
        
        diff = torch.mean((epoch_checkpoint[param_name].float().flatten() - end_checkpoint[param_name].float().flatten())**2).sqrt()
        
#         print(epoch,param_name, diff)
        
        
        results[param_name] = results.get(param_name,[])
        
        results[param_name].append(diff)
        
p["results"][p["model_name"]] = results
```

```python


results = {}
for epoch in range(p["end_epoch"]):
    print(epoch)
    e0 = load_epoch(epoch,p)
    e1 = load_epoch(epoch+1,p)
    
    for param_name in e1.keys():
        
        diff = torch.mean((e0[param_name].float().flatten() - e1[param_name].float().flatten())**2).sqrt()
        
#         print(epoch,param_name, diff)
        
        
        results[param_name] = results.get(param_name,[])
        
        results[param_name].append(diff)
        
p["results"][p["model_name"]] = results
```

```python
results = p["results"][p["model_name"]]


layer_names = [name for name in results.keys() if "conv" in name]
print(layer_names)
num_layers = len(layer_names)

num_epochs = len(results[layer_names[0]])

print(num_layers,num_epochs)

img = np.zeros((num_layers,num_epochs))

for i_layer,layer_name in enumerate(layer_names[1:]):
    for i_epoch in range(num_epochs):
        diff = results[layer_name][i_epoch]
        img[i_layer,i_epoch] = diff
        
        
plt.figure()
plt.imshow(img)
        
```

```python
plt.figure()
for i in range(100):
    plt.plot(img.T[i])
```
